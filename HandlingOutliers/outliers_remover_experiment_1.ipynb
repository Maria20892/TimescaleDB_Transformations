```python
import pandas as pd
import numpy as np

# Function to detect outliers based on IQR
def detect_outliers(data):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    # Calculate the Interquartile Range (IQR)
    IQR = Q3 - Q1
    # Calculate lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Identify outliers
    outliers = [x for x in data if x < lower_bound or x > upper_bound]
    return outliers, lower_bound, upper_bound

# Load your Excel file
file_path = 'DraftToIdentifyOutliers.xlsx' 
df = pd.read_excel(file_path)

column_data = df.iloc[:, 3]

outliers_result = []

# Process each block of 5 rows
for i in range(0, len(column_data), 5):
    block = column_data[i:i+5]
    outliers, lower_bound, upper_bound = detect_outliers(block)
    if outliers:
        outliers_result.append({
            'Block': f'Rows {i+1}-{i+5}',
            'Outliers': outliers,
            'Lower Bound': lower_bound,
            'Upper Bound': upper_bound
        })

# Print the outliers result
for result in outliers_result:
    print(f"Block: {result['Block']}")
    print(f"Outliers: {result['Outliers']}")
    print(f"Lower Bound: {result['Lower Bound']}")
    print(f"Upper Bound: {result['Upper Bound']}")
    print("-" * 40)

# Remove outliers from the dataframe
filtered_df = df[~df.iloc[:, 3].isin([outlier for result in outliers_result for outlier in result['Outliers']])]

# Save the filtered dataframe without outliers
filtered_df.to_excel('filtered_dataColumnLOCF.xlsx', index=False)

print("\nOutliers removed and file saved as 'filtered_data.xlsx'")
```
