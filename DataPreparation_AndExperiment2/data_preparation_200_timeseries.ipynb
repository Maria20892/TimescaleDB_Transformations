Create 200 csv files from Exp3P1V5happiness_with_gaps.csv 

```python
import shutil
import os

# Function to create copies of a file with numbered names
def create_file_copies(original_file, destination_directory, num_copies):
    for i in range(1, num_copies + 1):
        # Generate the new file name
        new_file_name = f"{destination_directory}/Exp3P1V5happiness_with_gaps_{i}.csv"
        # Copy the original file to the new file name
        shutil.copy(original_file, new_file_name)
        print(f"Created: {new_file_name}")

# Specify the path to the original file
original_file_path = "Exp3P1V5happiness_with_gaps.csv"  
destination_directory = "./copies"
number_of_copies=200

if not os.path.exists(destination_directory):
    os.makedirs(destination_directory)

# Create copies of the file
create_file_copies(original_file_path, destination_directory, number_of_copies)

```
